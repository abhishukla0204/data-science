```python
# Load iris dataset (comes with seaborn)
iris = sns.load_dataset('iris')

# Create horizontal boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='sepal_length', y='species', data=iris, 
            palette='pastel', 
            showfliers=True, 
            flierprops=dict(marker='x', color='red', markersize=8))

plt.title('BoxPlot')
plt.xlabel('sepal length')
plt.ylabel('species')

# Remove legend (species is already on y-axis)
plt.legend([],[], frameon=False)

plt.show()
```
